```python
import numpy as np
from scipy.special import logsumexp
import matplotlib.image as mpimg
from preprocessing.Optimal_crop import optimal_crop
import matplotlib.pyplot as plt
```

```python
class GaussianMixtureModel:
    """
    A Gaussian mixture model to represent a provided grayscale image.
    """

    def __init__(self, image_matrix, num_components,error, iter_n ,means=None):
        """
        Initialize a Gaussian mixture model.
        params:
        image_matrix = (grayscale) numpy.nparray[numpy.nparray[float]]
        num_components = int
        """
        self.image_matrix = image_matrix
        self.num_components = num_components
        if means is None:
            self.means = np.zeros(num_components)
        else:
            self.means = means
        self.error=error
        self.iter_n =iter_n
        self.variances = np.zeros(num_components)
        self.mixing_coefficients = np.zeros(num_components)

    def joint_prob(self, val):
        """Calculate the joint log probability of a grayscale value within the image.
        ln( N( x|mu, sigma)) = -0.5 ln(2 x PI x sigma sqr) - (( val - mean ) sqr / 2 x sigma sqr)
        params:
        val = float
        returns:
        joint_prob = float
        """
        return logsumexp (np.log(self.mixing_coefficients)
            + (-0.5 * np.log(2 * np.pi * np.square(self.variances)) - (np.square(val - self.means)/ (2 * np.square(self.variances)))))

    def initialize_training(self):
        """
        Initialize the training process by setting each component mean to a random pixel's value (without replacement),
        each component variance to 1, and each component mixing coefficient to a uniform value (e.g. 4 components -> [0.25,0.25,0.25,0.25]).
        """
        # # print "SHAPE: ", self.image_matrix.shape
        [row_count, column_count] = self.image_matrix.shape
        image_matrix_reshaped = self.image_matrix.reshape(row_count * column_count, 1)
        self.means = np.squeeze(image_matrix_reshaped[np.random.choice(np.arange(row_count * column_count), self.num_components, replace = False)])
        self.variances = np.ones(self.num_components)
        self.mixing_coefficients = np.ones(self.num_components)/self.num_components

    def convergence_function(self,iter_n,error,conv_ctr,prev_likelihood, curr_likelihood):
        if (abs(prev_likelihood - curr_likelihood) > error) and (iter_n != conv_ctr) :
            return False
        else:
            return True

    def train_model(self):
        """
        Train the mixture model using the expectation-maximization algorithm. Since each Gaussian is a combination of mean and variance, this will fill self.means and self.variances, plus self.mixing_coefficients, with the values that maximize the overall model likelihood.
        params:
        convergence_function = function, returns True if convergence is reached
        """
        prev_likelihood = self.likelihood()
        conv_ctr = 0
        while True:
            [row_count, column_count] = self.image_matrix.shape
            image_array = self.image_matrix.reshape(row_count * column_count, 1)
            image_array_axised = image_array[:, np.newaxis]

            variances_axised = self.variances[:, np.newaxis]
            log_mixing_coefficients = np.log(self.mixing_coefficients[:, np.newaxis])

            deviation = image_array_axised - self.means[:, np.newaxis]
            logNormalDist = -0.5 * np.log( 2 * np.pi * variances_axised) - (np.square(deviation) / (2 * variances_axised))

            logGamma = np.add(logNormalDist, log_mixing_coefficients) - logsumexp(np.add(logNormalDist, log_mixing_coefficients), axis=1)[:, np.newaxis]

            gamma = np.exp(logGamma)
            N = row_count * column_count

            self.mixing_coefficients = (np.squeeze(np.sum(gamma, axis = 0) / N))
            new_means_num = np.sum(gamma * image_array_axised, axis = 0)
            normz_factor = np.sum(gamma, axis=0)
            self.means = np.squeeze(new_means_num / normz_factor)
            deviation = image_array_axised - self.means[:, np.newaxis]

            sigmaNum = np.sum(gamma * np.square(deviation), axis = 0)
            self.variances = np.squeeze(sigmaNum / normz_factor)
            curr_likelihood = self.likelihood()
            conv_ctr +=1
            converged = self.convergence_function(self.iter_n,conv_ctr,self.error, prev_likelihood, curr_likelihood,)
            if converged:
                break

            prev_likelihood = curr_likelihood

    def segment(self):
        """
        Segments the image matrix into the pre-specified number of components.
        Returns the original image matrix with the each pixel's intensity replaced with its max-likelihood component mean.
        returns:
        segment = numpy.ndarray[numpy.ndarray[float]]
        """

        [row_count, column_count] = self.image_matrix.shape
        image_array = self.image_matrix.reshape(row_count * column_count, 1)
        image_array_axised = image_array[:, np.newaxis]
        means_axised = self.means[:, np.newaxis]

        #diff = image_array_axised - means_axised
        sums = np.square(image_array_axised - means_axised)

        min_indices = np.squeeze(np.argmin(sums, axis=1))

        new_image_values_reshaped = np.take(self.means, min_indices, axis =0 )
        return new_image_values_reshaped.reshape(row_count, column_count)

    def likelihood(self):
        """Assigns a log likelihood to the trained model based on the following formula for posterior probability:
        ln(Pr(X | mixing, mean, stdev)) = sum((n=1 to N), ln(   sum(  (k=1 to K), mixing_k * N(x_n | mean_k,stdev_k))))
        returns:
        log_likelihood = float [0,1]
        """

        if type(self.means) is list:
            self.means = np.array(self.means)

        [row_count, column_count] = self.image_matrix.shape
        image_array = self.image_matrix.reshape(row_count * column_count, 1)
        image_array_axised = image_array[:, np.newaxis]

        variances_axised = self.variances[:, np.newaxis]

        logNx = -0.5 * np.log(2 * np.pi * variances_axised) - ( np.square( image_array_axised - self.means[:, np.newaxis] ) / (2 * variances_axised))
        logMix = np.log(self.mixing_coefficients[:, np.newaxis])

        A = np.add(logNx, logMix)
        return np.sum(logsumexp(A, axis=1))

    def best_segment(self, iters):
        """
        Determines the best segmentation of the image by repeatedly training the model and calculating its likelihood.
        Return the segment with the highest likelihood.
        params:
        iters = int
        returns:
        segment = numpy.ndarray[numpy.ndarray[float]]
        """

        best_means = self.means
        best_likelihood = self.likelihood()

        for i in range(iters):
            curr_likelihood = self.likelihood()
            if curr_likelihood > best_likelihood:
                best_means = self.means
                best_likelihood = curr_likelihood
            self.train_model()

        self.means = best_means
        return self.segment()
```
